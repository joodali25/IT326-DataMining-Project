# Project Goal

The goal of this project is to apply classification and clustering techniques on a real-world student mental health dataset. The classification model aims to predict the depression status of students (Yes/No) based on academic, social, and lifestyle attributes such as age, academic pressure, study satisfaction, sleep duration, dietary habits, financial stress, and family history of mental illness. The clustering model aims to uncover patterns by grouping students with similar characteristics, providing insights into the relationships between various factors associated with student depression.

# Dataset Source

The dataset used in this project was obtained from Kaggle:

https://www.kaggle.com/datasets/ikynahidwin/depression-student-dataset


# Dataset Description

### 1. General Information
* **Number of Instances (Records):** 101,500
* **Number of Attributes (Features):** 19
* **Dataset Source:** [Kaggle - Depression Student Dataset](https://www.kaggle.com/datasets/ikynahidwin/depression-student-dataset)

### 2. Attribute Information
The dataset includes the following features:

| Attribute | Data type | Possible values |
| :--- | :--- | :--- |
| **Gender** | Categorical | Male or Female |
| **Age** | Numeric | Age of the individual |
| **City** | Categorical | Location of the individual |
| **Working Professional or Student** | Categorical | Current occupation status |
| **Academic Pressure** | Numeric (Scale) | Level of pressure from studies (1-5) |
| **Work Pressure** | Numeric (Scale) | Level of pressure from work (1-5) |
| **Study Satisfaction** | Numeric (Scale) | Level of satisfaction with studies (1-5) |
| **Job Satisfaction** | Numeric (Scale) | Level of satisfaction with job (1-5) |
| **Sleep Duration** | Categorical | Average hours of sleep per night |
| **Dietary Habits** | Categorical | Type of diet (Healthy, Moderate, Unhealthy) |
| **Degree** | Categorical | Educational qualification |
| **Suicidal thoughts?** | Categorical | Yes or No |
| **Work/Study Hours** | Numeric | Daily hours spent working or studying |
| **Financial Stress** | Numeric (Scale) | Level of financial stress (1-5) |
| **Family History of Mental Illness** | Categorical | Yes or No |
| **Depression (Class Label)** | Categorical (Binary) | **Target Variable:** Yes (1) or No (0) |
